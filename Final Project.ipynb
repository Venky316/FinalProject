```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

getpd = pd.read_csv('Influencer.csv')
getpd.columns = ['Rank', 'Channel Info', 'Influence Score', 'Followers (in Millions)', 'Avg. Likes (in Millions)', 'Posts (in Thousands)', '60-Day Eng Rate', 'New Post Avg. Likes (In Millions)', 'Total Likes (In Billions)', 'Country']

for i in getpd['Channel Info'].index:
    getval = getpd.loc[i,'Channel Info']
    searchval = re.findall('\n',getval)
    if(len(searchval) > 0):
        getpd.loc[i,'Channel Info'] = getval.replace('\n','')

for i in getpd['Followers (in Millions)'].index:
    getval = getpd.loc[i,'Followers (in Millions)']
    getpd.loc[i,'Followers (in Millions)'] = getval.replace('m','')
getpd['Followers (in Millions)'] = getpd['Followers (in Millions)'].astype('float64')

for i in getpd['Avg. Likes (in Millions)'].index:
    getval = getpd.loc[i,'Avg. Likes (in Millions)']
    searchvalm = re.findall('m',getval)
    searchvalk = re.findall('k',getval)
    if(len(searchvalm) > 0):
        getpd.loc[i,'Avg. Likes (in Millions)'] = getval.replace('m','')
    elif(len(searchvalk) > 0):
        getpd.loc[i,'Avg. Likes (in Millions)'] = round((float(getval.replace('k','')) / 1000),1)
getpd['Avg. Likes (in Millions)'] = getpd['Avg. Likes (in Millions)'].astype('float64')

for i in getpd['Posts (in Thousands)'].index:
    getval = getpd.loc[i,'Posts (in Thousands)']
    searchvalm = re.findall('k',getval)
    if(len(searchvalm) > 0):
        getpd.loc[i,'Posts (in Thousands)'] = getval.replace('k','')
    else:
        getpd.loc[i,'Posts (in Thousands)'] = round((float(getval) / 1000),1)
getpd['Posts (in Thousands)'] = getpd['Posts (in Thousands)'].astype('float64')

for i in getpd['New Post Avg. Likes (In Millions)'].index:
    getval = getpd.loc[i,'New Post Avg. Likes (In Millions)']
    searchvalm = re.findall('m',getval)
    searchvalk = re.findall('k',getval)
    if(len(searchvalm) > 0):
        getpd.loc[i,'New Post Avg. Likes (In Millions)'] = getval.replace('m','')
    elif(len(searchvalk) > 0):
        getpd.loc[i,'New Post Avg. Likes (In Millions)'] = round((float(getval.replace('k','')) / 1000),1)
getpd['New Post Avg. Likes (In Millions)'] = getpd['New Post Avg. Likes (In Millions)'].astype('float64')

for i in getpd['Total Likes (In Billions)'].index:
    getval = getpd.loc[i,'Total Likes (In Billions)']
    searchvalb = re.findall('b',getval)
    searchvalm = re.findall('m',getval)
    if(len(searchvalb) > 0):
        getpd.loc[i,'Total Likes (In Billions)'] = getval.replace('b','')
    elif(len(searchvalm) > 0):
        getpd.loc[i,'Total Likes (In Billions)'] = round((float(getval.replace('m','')) / 1000),1)
getpd['Total Likes (In Billions)'] = getpd['Total Likes (In Billions)'].astype('float64')

getpd = getpd.iloc[:,[0,1,9,2,3,4,5,6,7,8]]

copypd = getpd.copy()
copypd.drop(['Rank','Channel Info','Country'],axis=1,inplace=True)

scaler = StandardScaler()
copypd = pd.DataFrame(scaler.fit_transform(copypd))
copypd.columns = ['Influence Score', 'Followers (in Millions)', 'Avg. Likes (in Millions)', 'Posts (in Thousands)', '60-Day Eng Rate', 'New Post Avg. Likes (In Millions)', 'Total Likes (In Billions)']

# Question 1

copypd.corr()

# Question 2

sns.histplot(copypd['Influence Score'],kde=True)
sns.histplot(copypd['Followers (in Millions)'],kde=True)
sns.histplot(copypd['Posts (in Thousands)'],kde=True)

# Question 3

getcountlist = getpd['Country'].unique().tolist()
getlist = []

for i in getcountlist:
    getval = getpd[getpd['Country'] == i].shape[0]
    getlist.append([i,getval])

newpd = pd.DataFrame(getlist,columns=['Country','No. of Influencers'])
plt.figure(figsize=(12,12))
sns.barplot(newpd,x='No. of Influencers',y='Country')

# Question 4

testpd = getpd.sort_values('Followers (in Millions)',ascending=False)
testpd[['Channel Info','Followers (in Millions)']].head(10)

testpd = getpd.sort_values('Avg. Likes (in Millions)',ascending=False)
testpd[['Channel Info','Avg. Likes (in Millions)']].head(10)

testpd = getpd.sort_values('Total Likes (In Billions)',ascending=False)
testpd[['Channel Info','Total Likes (In Billions)']].head(10)

# Question 5

plt.figure(figsize=(15,15))
sns.lmplot(getpd,x='Followers (in Millions)',y='Total Likes (In Billions)')
sns.lineplot(getpd,x='Followers (in Millions)',y='Total Likes (In Billions)')

plt.figure(figsize=(15,15))
sns.lmplot(getpd,x='Followers (in Millions)',y='Influence Score')
sns.lineplot(getpd,x='Followers (in Millions)',y='Influence Score')

plt.figure(figsize=(15,15))
sns.lmplot(getpd,x='Posts (in Thousands)',y='Avg. Likes (in Millions)')
sns.lineplot(getpd,x='Posts (in Thousands)',y='Avg. Likes (in Millions)')

plt.figure(figsize=(15,15))
sns.lmplot(getpd,x='Posts (in Thousands)',y='Influence Score')
sns.lineplot(getpd,x='Posts (in Thousands)',y='Influence Score')
```
